```python
import base64
import os
from langchain_core.messages import HumanMessage
```

```python
def encode_image(image_path):
    """Getting the base64 string"""
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
```

```python
from langchain_huggingface import HuggingFaceEndpoint

llm = HuggingFaceEndpoint(
    repo_id="Salesforce/blip-image-captioning-base",
    task="image-to-text"
)

def image_summarize(img_base64, prompt):
    msg = llm.invoke([
        HumanMessage(
            content=[
                {"type": "text", "text": prompt},
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:image/jpeg;base64,{img_base64}"},
                },
            ]
        )
    ])
    return msg.content


def generate_img_summaries(path):
    img_base64_list = []
    image_summaries = []

    prompt = """
    Summarize the image for semantic retrieval.
    Include objects, actions, text, and context.
    """

    files = [path] if os.path.isfile(path) else sorted(os.listdir(path))

    for img_file in files:
        if not img_file.lower().endswith((".jpg", ".jpeg", ".png")):
            continue

        img_path = path if os.path.isfile(path) else os.path.join(path, img_file)

        try:
            base64_image = encode_image(img_path)
            summary = image_summarize(base64_image, prompt)

            img_base64_list.append(base64_image)
            image_summaries.append(summary)

        except Exception as e:
            print(f"Error processing {img_file}: {e}")

    return img_base64_list, image_summaries
```
