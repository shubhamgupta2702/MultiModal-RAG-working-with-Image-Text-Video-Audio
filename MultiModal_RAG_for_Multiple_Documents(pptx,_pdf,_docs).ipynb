```python
from langchain_community.document_loaders import DirectoryLoader
from langchain_huggingface import ChatHuggingFace, HuggingFaceEmbeddings, HuggingFaceEndpoint
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
```

```python
loader = DirectoryLoader(
    path="/content/docs",
)
```

```python
docs = loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=150, chunk_overlap=50)
```

```python
docs = splitter.split_documents(docs)
```

```python
len(docs)
```

```python
import os
from google.colab import userdata
HUGGINGFACEHUB_API_TOKEN=userdata.get('HUGGINGFACEHUB_API_TOKEN')
os.environ["HUGGINGFACEHUB_API_TOKEN"] = HUGGINGFACEHUB_API_TOKEN
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
from langchain_astradb import AstraDBVectorStore
```

```python
from google.colab import userdata
ASTRA_DB_APPLICATION_TOKEN=userdata.get('ASTRA_DB_APPLICATION_TOKEN')
ASTRA_DB_API_ENDPOINT=userdata.get('ASTRA_DB_API_ENDPOINT')
ASTRA_DB_KEYSPACE = "default_keyspace"
```

```python
vector_store = AstraDBVectorStore(
    embedding=embeddings,
    collection_name="my_collection",
    api_endpoint=ASTRA_DB_API_ENDPOINT,
    token=ASTRA_DB_APPLICATION_TOKEN,
    namespace=ASTRA_DB_KEYSPACE
)
```

```python
inserted_docs = vector_store.add_documents(docs)
```

```python
len(inserted_docs)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k":3})
```

```python
prompt = PromptTemplate(
    template="""
    You are an AI Philospher drawing insights from the documents related to transformers, llms, AI ML, sherlock homes stories, RAG, It lab projects.
    With this answer to the user query efficiently.
    DOCUMENT_CONTEXT: \n
    {context}\n

    Query:\n
    {query}
    """,
    input_variables=['context', 'query']
)
```

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough
parser = StrOutputParser()
```

```python
llm = HuggingFaceEndpoint(
  repo_id="Qwen/Qwen3-Coder-Next",
  task='text-generation'
)

model = ChatHuggingFace(llm=llm)
```

```python
chain = (
    {"context":retriever, "query":RunnablePassthrough()} | prompt | model | parser
)
```

```python
chain.invoke("What is RAG")
```
